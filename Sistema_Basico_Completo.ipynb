* **Na opção 3**, o valor da compra, deve ser descontado no saldo do cashback do cliente que a realizou. Por exemplo, cliente 1, tem 500 de saldo, se ele realizar uma compra de 100 reais, o saldo do cashback passará para 400.
* **Na opção 5**, o usuário do sistema informará/digitará o código do cliente que quer e o sistema/programa fará uma pesquisa e calculará o total de compras do determinado cliente.

```
dadosCliente = []
dadosCompra = []
class Cliente:
    id = 0
    nome = ''
    cash = 0.0
class Compra:
    idComp = 0
    id = 0
    valor = 0.0
def cadastroCliente():#1
    newCliente = Cliente()
    newCliente.id = int(input('Informe o ID do CLIENTE: '))
    if existeCliente(newCliente.id) == False:
        newCliente.nome = input('Informe o NOME do cliente: ')
        newCliente.cash = float(input('Informe o saldo do CASHBACK do cliente: '))
        dadosCliente.append(newCliente)
        print('\n\tCliente cadastrado.')
    else:
        print('\n\tCliente não encontrado.')
def existeCliente(id):
    for i in range(len(dadosCliente)):
        if dadosCliente[i].id == id:
            return True
    return False
def exibeClientes():#2
    print('ID\tNOME\t     CASH')
    for i in range(len(dadosCliente)):
        print(f'{dadosCliente[i].id:<8}{dadosCliente[i].nome[:10]:<13}R${dadosCliente[i].cash:.2f}')
def posicaoCliente(id):
    for i in range(len(dadosCliente)):
        if dadosCliente[i].id == id:
            return i
def cadastroCompra():#3
    newCompra = Compra()
    newCompra.idComp = int(input('Informe o ID da COMPRA: '))
    if existeIdCompra(newCompra.idComp) == False:
        newCompra.id = int(input('Informe o ID do CLIENTE: '))
        if existeCliente(newCompra.id):
            newCompra.valor = float(input('Informe o VALOR da compra: '))
            dadosCliente[posicaoCliente(newCompra.id)].cash = dadosCliente[posicaoCliente(newCompra.id)].cash - newCompra.valor
            dadosCompra.append(newCompra)
            print('\n\tCompra efetuada.')
        else:
            print('\n\tCliente não encontrado.')
    else:
        print('\n\tCódigo de compra já existe.')
def existeIdCompra(idComp):
    for i in range(len(dadosCompra)):
        if dadosCompra[i].idComp == idComp:
            return True
    return False
def exibeCompra():#4
    print('ID COMP.  ID CLI.   VALOR')
    for i in range(len(dadosCompra)):
        print(f'{dadosCompra[i].idComp:<10}{dadosCompra[i].id:<10}R${dadosCompra[i].valor:.2f}')
def totalCompra_cliente():#5
    id = int(input('Informe o ID do CLIENTE para buscar: '))
    if existeCliente(id):
        class Soma: #criado um objeto para poder somar outros objetos, pois uma variavel não recebe soma de objetos
            valor = 0.0
        soma = Soma()
        cont = 0
        for i in range(len(dadosCompra)):
            if dadosCompra[i].id == id:
                soma.valor = dadosCompra[i].valor + soma.valor
                cont += 1
        print(f'\nREF: ID {id}.\nFez {cont} compra(s).\nSomando R${soma.valor:.2f}.') 
    else:
        print('\n\tCliente não encontrado.')
def salvarDados():#6
    arquivo = open('compras.csv','w')
    idComp, id, valor = 'COMP','CLIN','VALOR'
    arquivo.write(f'{idComp},{id},{valor}\n')
    for i in range(len(dadosCompra)):
        arquivo.write(f'{dadosCompra[i].idComp},{dadosCompra[i].id},{dadosCompra[i].valor}\n')
    arquivo.close()
    print('\n\tDados Salvos.')
def exibeDados():#7
    arquivo = open('compras.csv','r')
    for linha in arquivo.readlines():
        idComp, id, valor = linha.strip().split(',')
        print(f'{idComp}\t{id}\t{valor}')
    arquivo.close()
def exibeGrafico():#8
    import pandas as pd
    df1 = pd.read_csv('compras.csv')
    import matplotlib.pyplot as plt
    plt.hist(df1['VALOR'], 5, rwidth = 0.75, color = 'orange')
    plt.title('Resumo de vendas')
    plt.xlabel('Preço')
    plt.ylabel('Frequência')
    plt.show()
def sair():
    return False
def opcoesDisponiveis(opcao):
    if opcao == 1:
        cadastroCliente()
    elif opcao == 2:
        exibeClientes()
    elif opcao == 3:
        cadastroCompra()
    elif opcao == 4:
        exibeCompra()
    elif opcao == 5:
        totalCompra_cliente()
    elif opcao == 6:
        salvarDados()
    elif opcao == 7:
        exibeDados()
    elif opcao == 8:
        exibeGrafico()
def validaOpcoes(opcao):
    if opcao > 0 and opcao < 10:
        return True
    else:
        return False
def menu():
    print('\nMenu de Opções\n1 Cadastrar clientes\n2 Mostrar todos os clientes\n3 Cadastrar compras\n4 Mostrar todas as compras\n5 Mostrar o total de todas as compras realizadas por um cliente\n6 Armazenar todas as compras em arquivo\n7 Apresentar o conteúdo do arquivo\n8 Gerar um gráfico\n9 Sair\n')
def main():
    validacao = True
    while validacao:
        menu()
        opcao = int(input('Informe a opção desejada: '))
        if validaOpcoes(opcao):
            if opcao == 9:
                print('\tVocê escolheu sair...')
                validacao = sair()
            opcoesDisponiveis(opcao)
        else:
            print('\n\tOpção inválida.\n')
main()
```
